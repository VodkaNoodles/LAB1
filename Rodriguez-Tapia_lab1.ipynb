```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading the database
students = pd.read_csv("student-mat.csv", sep = ';', quotechar=" ")
print(students.head())
```

```python


average_freetime_by_age = students.groupby('age')['freetime'].mean()
print("\nAverage Free Time by Age")
print(average_freetime_by_age)



plt.figure(figsize=(8,6))
average_freetime_by_age.plot(kind='line')
plt.xlabel('Age')
plt.ylabel('Freetime')
plt.title('Average Free Time by Age')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
