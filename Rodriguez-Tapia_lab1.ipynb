```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

students = pd.read_csv("student-mat.csv", sep = ';', quotechar=" ")\
#display
print(students.head())
```

Heres is the bar graph, this shows the comparison between students who're in a relationship and the average number of absences. As shown in the graph students who're in a relationship have on average more absences compared to students who are not in a relationship

```python
#separate students by relationship status
romantic = students[students['romantic'] == 'yes']
non_romantic = students[students['romantic'] == 'no']

#calculate average absence by group (in relationship or not)
average_absences_by_romantic_status = students.groupby('romantic')['absences'].mean()
print("\nAverage skip and romantic status:")
print(average_absences_by_romantic_status)



plt.figure(figsize=(8,6))
average_absences_by_romantic_status.plot(kind='bar')
plt.xlabel('In Relationship')
plt.ylabel('Absences')
plt.title('Average Absences v Student Romantic Status')
plt.grid(axis='y', linestyle='--', alpha=0.7)
#save figure as png
plt.savefig("SaulRodriguez-Tapia_BarGraph.png")
plt.show()
```

This is the line graph, the Y-axis is on a scale of 3 (least free time) to 5 (most free time), this shows the relationship between age and free time the students have, as shown there's a slight gradual decrease in free time from 15 to 18 where 18 year old students have the least free time. Then there is a significant increase in student free time from 19 to about 19.5 then there is a decrease in free time starting at 21

```python
#get average free time per age
average_freetime_by_age = students.groupby('age')['freetime'].mean()
print("\nAverage Free Time by Age")
print(average_freetime_by_age)



plt.figure(figsize=(8,6))
average_freetime_by_age.plot(kind='line')
plt.xlabel('Age')
plt.ylabel('Freetime')
plt.title('Average Free Time by Age')
plt.grid(axis='y', linestyle='--', alpha=0.7)
#save figure as png
plt.savefig("SaulRodriguez-Tapia_LineGraph.png")
plt.show()
```

This pie chart shows all the students that have or dont have internet access, as shown more than 80% of students do have internet acess and about 16% do not have internet access

```python
# Count the number of students with and without internet
internet_counts = students['internet'].value_counts()

# Plot the pie chart
plt.figure(figsize=(6,6))
plt.pie(internet_counts, 
        labels=internet_counts.index, 
        autopct='%1.1f%%', 
        startangle=90,
        shadow=True)

plt.title("Internet Access Among Students")
plt.axis('equal') 
#save figure as png
plt.savefig("SaulRodriguez-Tapia_PieChart.png")
plt.show()
```

This is the box plot, this graph shows the average final grades of 4 categories of students 1-4 1 being students who studied less than 2 hours to 4 being students who studied more than 10 hours, this shows the high and low outliers along with the average per category

```python
#use G3 column for box plot
students.boxplot(column='G3', by='studytime')

plt.title("Final Grades (G3) by Study Time")
plt.suptitle("")
plt.xlabel("Study Time (1 = <2h, 4 = 10+h)")
plt.ylabel("Final Grade (G3)")
plt.grid(True, linestyle='--', alpha=0.6)
#save figure as png
plt.savefig("SaulRodriguez-Tapia_BoxChart.png")
plt.show()
```
