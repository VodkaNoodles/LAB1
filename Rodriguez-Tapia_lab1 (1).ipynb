```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading the database
students = pd.read_csv("student-mat.csv", sep = ';', quotechar=" ")
print(students.head())
```

```python
romantic = students[students['romantic'] == 'yes']
non_romantic = students[students['romantic'] == 'no']

average_absences_by_romantic_status = students.groupby('romantic')['absences'].mean()
print("\nAverage skip and romantic status:")
print(average_absences_by_romantic_status)



plt.figure(figsize=(8,6))
average_absences_by_romantic_status.plot(kind='bar')
plt.xlabel('Romantic Status')
plt.ylabel('Absences')
plt.title('Average Absences v Student Romantic Status')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
average_freetime_by_age = students.groupby('age')['freetime'].mean()
print("\nAverage Free Time by Age")
print(average_freetime_by_age)



plt.figure(figsize=(8,6))
average_freetime_by_age.plot(kind='line')
plt.xlabel('Age')
plt.ylabel('Freetime')
plt.title('Average Free Time by Age')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
